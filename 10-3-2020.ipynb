```python
# i/p : [27,24,12,19,30]
# o/p : [27,576,144,19,900]
a = [27,24,12,19,30]
for i in range(len(a)):
    if (a[i] % 6) ==0:
        a[i] =a[i]**2
a
```

```python
a = [27,24,12,19,30]
b = []
for i in range(len(a)):
    if (a[i] % 6) ==0:
        b.append(a[i]**2)
    else:
         b.append(a[i])
a
```

```python
a = [27,24,12,19,30]
b = []
for i in a:
    if (i % 6) ==0:
        b.append(i**2)
    else:
         b.append(i)
b
```

```python
a = [27,24,12,19,30]
b = a
for i in range(len(a)):
    if (a[i] % 6) ==0:
        b[i] = b[i]**2
print(a)
print(b)
```

```python
a = [27,24,12,19,30]
b = a.copy()
for i in range(len(a)):
    if (a[i] % 6) ==0:
        b[i] = b[i]**2
print(a)
print(b)
```

```python
a = [27,24,12,19,30]
b = a[:]
for i in range(len(a)):
    if (a[i] % 6) ==0:
        b[i] = b[i]**2
print(a)
print(b)
```
